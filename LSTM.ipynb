```python
# ==========================================================
# Importation de la bibliothèque NumPy
# ==========================================================
# NumPy permet de manipuler des tableaux et de faire des calculs matriciels.
import numpy as np

# ==========================================================
# FONCTIONS D'ACTIVATION
# ==========================================================

# Fonction sigmoïde
# Elle retourne une valeur comprise entre 0 et 1.
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Fonction tangente hyperbolique
# Elle retourne une valeur comprise entre -1 et 1.
def tanh(x):
    return np.tanh(x)

# ==========================================================
# DONNÉES D'ENTRÉE
# ==========================================================

# Séquence d'entrée composée de trois valeurs.
X = np.array([
    [1],
    [2],
    [3]
])

# ==========================================================
# PARAMÈTRES DU RÉSEAU
# ==========================================================

# Nombre de variables d'entrée.
input_size = 1

# Nombre de neurones cachés.
hidden_size = 2

# ==========================================================
# INITIALISATION DES POIDS
# ==========================================================

# Chaque porte reçoit :
# l'entrée x
# +
# l'ancien état caché h

# La taille devient donc :
# hidden_size × (input_size + hidden_size)

Wf = np.random.rand(hidden_size, input_size + hidden_size)   # Porte Forget
Wi = np.random.rand(hidden_size, input_size + hidden_size)   # Porte Input
Wo = np.random.rand(hidden_size, input_size + hidden_size)   # Porte Output
Wc = np.random.rand(hidden_size, input_size + hidden_size)   # Mémoire candidate

# ==========================================================
# BIAIS
# ==========================================================

bf = np.zeros((hidden_size,1))
bi = np.zeros((hidden_size,1))
bo = np.zeros((hidden_size,1))
bc = np.zeros((hidden_size,1))

# ==========================================================
# ÉTATS INITIAUX
# ==========================================================

# Etat caché initial
h = np.zeros((hidden_size,1))

# Etat mémoire initial
C = np.zeros((hidden_size,1))

# ==========================================================
# PROPAGATION AVANT
# ==========================================================

# Lire la séquence une valeur à la fois.
for t in range(len(X)):

    # Transformer l'entrée en vecteur colonne.
    x = X[t].reshape(input_size,1)

    # Concaténer l'entrée actuelle
    # avec l'ancien état caché.
    combined = np.vstack((h,x))

    # ======================================================
    # PORTE FORGET
    # ======================================================

    # Décider quelles informations oublier.
    f = sigmoid(np.dot(Wf,combined)+bf)

    # ======================================================
    # PORTE INPUT
    # ======================================================

    # Décider quelles nouvelles informations mémoriser.
    i = sigmoid(np.dot(Wi,combined)+bi)

    # ======================================================
    # MÉMOIRE CANDIDATE
    # ======================================================

    # Calculer les nouvelles informations possibles.
    C_tilde = tanh(np.dot(Wc,combined)+bc)

    # ======================================================
    # MISE À JOUR DE LA MÉMOIRE
    # ======================================================

    # Nouvelle mémoire =
    # ancienne mémoire conservée
    # +
    # nouvelles informations
    C = f*C + i*C_tilde

    # ======================================================
    # PORTE OUTPUT
    # ======================================================

    # Déterminer ce qui sera envoyé en sortie.
    o = sigmoid(np.dot(Wo,combined)+bo)

    # ======================================================
    # NOUVEL ÉTAT CACHÉ
    # ======================================================

    # Calcul de la sortie cachée.
    h = o*tanh(C)

    # ======================================================
    # AFFICHAGE
    # ======================================================

    print("--------------------------")
    print("Temps :",t+1)

    print("Entrée :")
    print(x)

    print("Forget Gate")
    print(f)

    print("Input Gate")
    print(i)

    print("Mémoire Candidate")
    print(C_tilde)

    print("Etat Mémoire")
    print(C)

    print("Output Gate")
    print(o)

    print("Etat Caché")
    print(h)
```
