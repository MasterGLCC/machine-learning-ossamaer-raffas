```python
import numpy as np      # bibliothèque pour calcul numérique (matrices)
import random           # pour exploration aléatoire

class QLearning:
    def __init__(self, states, actions, alpha=0.1, gamma=0.9, episodes=1000):
        # nombre d'états possibles dans l'environnement
        self.states = states

        # nombre d'actions possibles
        self.actions = actions

        # taux d'apprentissage (learning rate)
        self.alpha = alpha

        # facteur de réduction du futur (discount factor)
        self.gamma = gamma

        # nombre d'épisodes d'entraînement
        self.episodes = episodes

        # Q-table initialisée à zéro (états × actions)
        self.Q = np.zeros((states, actions))

    def reward(self, state):
        # fonction de récompense simple
        # si on atteint l'état final → récompense positive
        if state == self.states - 1:
            return 10
        # sinon pas de récompense
        return 0

    def update(self, state, action, reward, next_state):
        # mise à jour de la Q-table selon la formule Q-learning

        self.Q[state, action] = self.Q[state, action] + self.alpha * (
            reward + self.gamma * np.max(self.Q[next_state]) - self.Q[state, action]
        )

    def fit(self):
        # boucle d'entraînement sur plusieurs épisodes
        for episode in range(self.episodes):

            # état initial
            state = 0

            # tant qu'on n'a pas atteint l'état final
            while state != self.states - 1:

                # exploration : choisir une action aléatoire
                action = random.randint(0, self.actions - 1)

                # transition vers nouvel état
                if action == 1:
                    next_state = min(state + 1, self.states - 1)
                else:
                    next_state = max(state - 1, 0)

                # calcul de la récompense
                r = self.reward(next_state)

                # mise à jour de la Q-table
                self.update(state, action, r, next_state)

                # passage à l'état suivant
                state = next_state

    def predict(self, state):
        # retourne la meilleure action pour un état donné
        return np.argmax(self.Q[state])
```
