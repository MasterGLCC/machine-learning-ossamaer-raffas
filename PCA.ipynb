```python
import numpy as np

class PCA:
    def __init__(self, n_components):
        self.n_components = n_components
        self.mean = None
        self.components = None
        self.explained_variance = None

    def fit(self, X):
        # 1️Centrer les données (soustraire la moyenne)
        self.mean = np.mean(X, axis=0)
        X_centered = X - self.mean

        # 2️ Calculer la matrice de covariance
        # cov(X) = (1 / n) * X^T X
        cov_matrix = np.cov(X_centered, rowvar=False)

        # Calcul des valeurs propres et vecteurs propres
        eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

        # Trier par ordre décroissant des valeurs propres
        idx = np.argsort(eigenvalues)[::-1]
        eigenvalues = eigenvalues[idx]
        eigenvectors = eigenvectors[:, idx]

        # Sélectionner les k composantes principales
        self.components = eigenvectors[:, :self.n_components]

        # Variance expliquée
        self.explained_variance = eigenvalues[:self.n_components]

    def transform(self, X):
        # Centrer les données
        X_centered = X - self.mean

        # Projection sur les nouvelles dimensions
        return np.dot(X_centered, self.components)

    def fit_transform(self, X):
        self.fit(X)
        return self.transform(X)
```
