```python
import numpy as np  # importer la bibliothèque numpy pour les calculs mathématiques et les tableaux

# Données
X = np.array([1, 2, 3, 4, 5])  
# X représente les variables d'entrée

y = np.array([30000, 35000, 40000, 50000, 55000])  
# y représente les valeurs à prédire (exemple : salaire)



n = len(X)  # n = nombre d'observations (ici 5 données)

x_mean = np.mean(X)  # calcul de la moyenne de X

y_mean = np.mean(y)  # calcul de la moyenne de y


a = np.sum((X - x_mean) * (y - y_mean)) / np.sum((X - x_mean)**2)   # formule de la pente (coefficient directeur)
#



b = y_mean - a * x_mean  
# calcul de l'ordonnée à l'origine (valeur de y quand X = 0)

print("Pente a =", a)  # afficher la pente du modèle

print("Intercept b =", b)  # afficher l'interception du modèle


y_pred = a * X + b  # calcul des valeurs prédites par le modèle

print("Prédictions :", y_pred)  # afficher les résultats de prédiction
```

```python









X1 = np.array([1, 2, 3, 4, 5])        # feature 1 (ex: expérience)
X2 = np.array([20, 21, 22, 23, 24])   # feature 2 (ex: âge)

y = np.array([30000, 35000, 40000, 50000, 55000])  # salaire


X = np.column_stack((X1, X2))  #crier  matrice (n, 2)

n = len(y)


#  Ajouter la colonne de 1 pour l'intercept

X_b = np.c_[np.ones((n, 1)), X]  # (n, 3)


#  Formule analytique (Normal Equation)

# theta = (X^T X)^(-1) X^T y
theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)


b = theta[0]      # intercept
a1 = theta[1]     # coefficient 1
a2 = theta[2]     # coefficient 2

print("Intercept b =", b)
print("Coefficient a1 =", a1)
print("Coefficient a2 =", a2)


y_pred = X_b.dot(theta)

print("Prédictions :", y_pred)


# 7. Test nouvelle donnée

new_x1 = 6
new_x2 = 25

new_data = np.array([1, new_x1, new_x2])  # 1 pour intercept
prediction = new_data.dot(theta)

print("Prédiction nouveau salaire :", prediction)
```

```python


X = np.array([1, 2, 3, 4, 5])  # variable d'entrée (ex: x)

y = np.array([2, 8, 18, 32, 50])  # variable cible (ex: y = x^2 approximé)


degree = 2  # on choisit un polynôme de degré 2 : y = a0 + a1*x + a2*x²


X_poly = np.column_stack([X**i for i in range(degree + 1)])  # crée une matrice avec :
# X^0 (colonne de 1 pour l'intercept)
# X^1 (valeurs normales)
# X^2 (carrés de X)




theta = np.linalg.inv(X_poly.T.dot(X_poly)).dot(X_poly.T).dot(y)  # formule mathématique pour trouver les coefficients :
# θ = (XᵀX)^(-1) Xᵀ y
# elle permet de minimiser l'erreur


print("Coefficients :", theta)  # affiche les valeurs de a0, a1, a2


y_pred = X_poly.dot(theta)  # calcule les valeurs prédites par le modèle

print("Prédictions :", y_pred)  # affiche les résultats estimés


x_new = np.array([1, 6, 6**2])  # on transforme x=6 en forme polynomiale :
# [1, x, x²]

prediction = x_new.dot(theta)  # calcule la prédiction pour x = 6

print("Prédiction pour x=6 :", prediction)  # affiche le résultat final
```
