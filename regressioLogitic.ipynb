```python
import numpy as np  # importer numpy pour les calculs mathématiques


X = np.array([1, 2, 3, 4, 5])  # variable d'entrée (feature)

y = np.array([0, 0, 0, 1, 1])  # variable cible (classe binaire : 0 ou 1)


# . Initialisation des paramètres

w = 0  # poids du modèle (coefficient qui contrôle l'influence de X)

b = 0  # biais (valeur de départ quand X = 0)

lr = 0.1  # learning rate : vitesse d'apprentissage

epochs = 1000  # nombre d'itérations pour entraîner le modèle


def sigmoid(z):
    return 1 / (1 + np.exp(-z))  # transforme une valeur en probabilité entre 0 et 1


n = len(X)  # nombre d'exemples

for i in range(epochs):  # boucle d'entraînement (répétée plusieurs fois pour améliorer le modèle)

    # calcul de la prédiction
    z = w * X + b  
    # combinaison linéaire : w*x + b

    y_pred = sigmoid(z)  
    # application de la sigmoïde pour obtenir une probabilité

    # erreur entre prédiction et vraie valeur
    error = y_pred - y  
    # mesure de la différence entre sortie et réalité

    # calcul du gradient pour w
    dw = (1/n) * np.dot(error, X)  
    # direction pour corriger w

    # calcul du gradient pour b
    db = (1/n) * np.sum(error)  
    # direction pour corriger b

    # mise à jour des paramètres
    w = w - lr * dw  
    # correction du poids

    b = b - lr * db  
    # correction du biais


print("Poids w =", w)  
# affiche le coefficient appris

print("Biais b =", b)  
# affiche l'intercept appris


def predict(x):
    z = w * x + b  
    # calcul linéaire

    return sigmoid(z)  
    # retourne une probabilité


for x in X:
    print(f"x={x} -> prob={predict(x):.3f}")  
# affiche la probabilité de classe 1 pour chaque x
```
