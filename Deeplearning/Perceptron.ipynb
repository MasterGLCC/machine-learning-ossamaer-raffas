```python
# Importation de la bibliothèque NumPy pour effectuer des calculs numériques
import numpy as np

# --------------------------------------------------
# Données d'entraînement (porte logique AND)
# --------------------------------------------------

# X contient les entrées du perceptron
# Chaque ligne représente un exemple d'apprentissage
X = np.array([
    [0, 0],   # Exemple 1
    [0, 1],   # Exemple 2
    [1, 0],   # Exemple 3
    [1, 1]    # Exemple 4
])

# y contient les sorties attendues
# Pour la porte AND, seule l'entrée [1,1] produit 1
y = np.array([0, 0, 0, 1])

# --------------------------------------------------
# Initialisation des paramètres
# --------------------------------------------------

# Initialisation des poids à zéro
# Il y a deux poids car il y a deux variables d'entrée
poids = np.zeros(2)

# Initialisation du biais à zéro
biais = 0

# Définition du taux d'apprentissage
# Il contrôle la vitesse d'apprentissage
taux_apprentissage = 0.1

# Nombre de passages sur toutes les données
epochs = 10

# --------------------------------------------------
# Fonction d'activation (fonction seuil)
# --------------------------------------------------

def activation(x):

    # Si la somme pondérée est positive ou nulle
    if x >= 0:
        return 1

    # Sinon retourner 0
    else:
        return 0

# --------------------------------------------------
# Début de l'apprentissage
# --------------------------------------------------

# Répéter l'entraînement pendant 10 époques
for epoch in range(epochs):

    # Afficher le numéro de l'époque
    print(f"\nEpoch {epoch+1}")

    # Parcourir tous les exemples d'entraînement
    for i in range(len(X)):

        # Calcul de la somme pondérée
        # z = x1*w1 + x2*w2 + biais
        z = np.dot(X[i], poids) + biais

        # Application de la fonction d'activation
        prediction = activation(z)

        # Calcul de l'erreur
        # erreur = valeur réelle - valeur prédite
        erreur = y[i] - prediction

        # Mise à jour des poids
        # Nouvelle valeur = ancienne valeur + correction
        poids = poids + taux_apprentissage * erreur * X[i]

        # Mise à jour du biais
        biais = biais + taux_apprentissage * erreur

        # Affichage des informations de chaque exemple
        print(f"Entrée : {X[i]}  Sortie : {prediction}  Erreur : {erreur}")

# --------------------------------------------------
# Affichage des paramètres appris
# --------------------------------------------------

# Afficher les poids obtenus après l'entraînement
print("\nPoids finaux :", poids)

# Afficher le biais final
print("Biais :", biais)

# --------------------------------------------------
# Phase de test
# --------------------------------------------------

print("\nPrédictions :")

# Tester le perceptron sur toutes les entrées
for x in X:

    # Calcul de la somme pondérée
    z = np.dot(x, poids) + biais

    # Affichage de la prédiction
    print(x, "->", activation(z))
```
