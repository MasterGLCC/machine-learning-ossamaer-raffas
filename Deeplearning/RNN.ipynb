```python
# Importation de la bibliothèque NumPy
# Elle permet d'effectuer des calculs sur les matrices.
import numpy as np

# ======================================================
# DONNÉES D'ENTRÉE
# ======================================================

# Chaque ligne représente une entrée de la séquence.
# Ici nous avons une séquence de 3 instants.
X = np.array([
    [1],
    [2],
    [3]
])

# ======================================================
# PARAMÈTRES DU RÉSEAU
# ======================================================

# Taille de l'entrée (une seule valeur)
input_size = 1

# Nombre de neurones cachés
hidden_size = 2

# Nombre de sorties
output_size = 1

# ======================================================
# INITIALISATION DES POIDS
# ======================================================

# Poids entre l'entrée et l'état caché
Wx = np.random.rand(hidden_size, input_size)

# Poids entre l'ancien état caché et le nouvel état caché
Wh = np.random.rand(hidden_size, hidden_size)

# Poids entre l'état caché et la sortie
Wy = np.random.rand(output_size, hidden_size)

# Biais de la couche cachée
bh = np.zeros((hidden_size,1))

# Biais de sortie
by = np.zeros((output_size,1))

# ======================================================
# FONCTION D'ACTIVATION TANH
# ======================================================

# Fonction tanh utilisée dans les RNN
def tanh(x):

    return np.tanh(x)

# ======================================================
# ÉTAT CACHÉ INITIAL
# ======================================================

# Au début il n'y a aucune mémoire.
h = np.zeros((hidden_size,1))

# ======================================================
# PROPAGATION AVANT
# ======================================================

# Parcourir chaque élément de la séquence
for t in range(len(X)):

    # Transformer l'entrée en vecteur colonne
    x = X[t].reshape(input_size,1)

    # Calcul du nouvel état caché
    h = tanh(
        np.dot(Wx,x)
        +
        np.dot(Wh,h)
        +
        bh
    )

    # Calcul de la sortie
    y = np.dot(Wy,h)+by

    # Afficher les résultats
    print("--------------------")
    print("Temps :",t+1)

    print("Entrée :")
    print(x)

    print("Etat caché :")
    print(h)

    print("Sortie :")
    print(y)
```
