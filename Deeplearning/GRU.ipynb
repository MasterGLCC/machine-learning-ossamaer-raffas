```python
# =====================================================
# Importation de la bibliothèque NumPy
# =====================================================
# NumPy permet de manipuler des matrices et de faire des calculs numériques.
import numpy as np

# =====================================================
# FONCTIONS D'ACTIVATION
# =====================================================

# Fonction Sigmoid
# Elle renvoie une valeur comprise entre 0 et 1.
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Fonction Tangente Hyperbolique
# Elle renvoie une valeur comprise entre -1 et 1.
def tanh(x):
    return np.tanh(x)

# =====================================================
# DONNÉES D'ENTRÉE
# =====================================================

# Séquence composée de trois valeurs.
X = np.array([
    [1],
    [2],
    [3]
])

# =====================================================
# PARAMÈTRES DU RÉSEAU
# =====================================================

# Nombre de variables d'entrée.
input_size = 1

# Nombre de neurones cachés.
hidden_size = 2

# =====================================================
# INITIALISATION DES POIDS
# =====================================================

# Poids de la porte Update.
Wz = np.random.rand(hidden_size, input_size + hidden_size)

# Poids de la porte Reset.
Wr = np.random.rand(hidden_size, input_size + hidden_size)

# Poids de l'état candidat.
Wh = np.random.rand(hidden_size, input_size + hidden_size)

# =====================================================
# INITIALISATION DES BIAIS
# =====================================================

bz = np.zeros((hidden_size,1))
br = np.zeros((hidden_size,1))
bh = np.zeros((hidden_size,1))

# =====================================================
# ÉTAT CACHÉ INITIAL
# =====================================================

# Au début, la mémoire est vide.
h = np.zeros((hidden_size,1))

# =====================================================
# PROPAGATION AVANT
# =====================================================

# Parcourir chaque élément de la séquence.
for t in range(len(X)):

    # Transformer l'entrée en vecteur colonne.
    x = X[t].reshape(input_size,1)

    # Concaténer l'entrée avec l'état caché précédent.
    combined = np.vstack((h, x))

    # =================================================
    # PORTE UPDATE
    # =================================================

    # Détermine combien d'informations anciennes conserver.
    z = sigmoid(np.dot(Wz, combined) + bz)

    # =================================================
    # PORTE RESET
    # =================================================

    # Détermine quelle partie de l'ancien état oublier.
    r = sigmoid(np.dot(Wr, combined) + br)

    # =================================================
    # ÉTAT CANDIDAT
    # =================================================

    # Combiner l'entrée avec l'état caché filtré.
    combined_reset = np.vstack((r * h, x))

    # Calcul du nouvel état candidat.
    h_tilde = tanh(np.dot(Wh, combined_reset) + bh)

    # =================================================
    # MISE À JOUR DE L'ÉTAT CACHÉ
    # =================================================

    # Calcul du nouvel état caché.
    h = (1 - z) * h + z * h_tilde

    # =================================================
    # AFFICHAGE
    # =================================================

    print("----------------------------")
    print("Temps :", t + 1)

    print("Entrée :")
    print(x)

    print("Update Gate :")
    print(z)

    print("Reset Gate :")
    print(r)

    print("Etat Candidat :")
    print(h_tilde)

    print("Etat Caché :")
    print(h)
```
