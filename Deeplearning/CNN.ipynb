```python
# Importation de la bibliothèque NumPy
# Elle permet de manipuler des tableaux et de faire des calculs matriciels.
import numpy as np

# ==========================================================
# IMAGE D'ENTRÉE
# ==========================================================

# Définition d'une image de taille 5×5.
# Chaque nombre représente l'intensité d'un pixel.
image = np.array([
    [1, 2, 3, 0, 1],
    [4, 5, 6, 1, 2],
    [7, 8, 9, 2, 3],
    [1, 2, 3, 4, 5],
    [6, 7, 8, 9, 0]
])

# ==========================================================
# FILTRE (KERNEL)
# ==========================================================

# Définition d'un filtre 3×3.
# Ce filtre est utilisé pour détecter les contours verticaux.
kernel = np.array([
    [ 1, 0,-1],
    [ 1, 0,-1],
    [ 1, 0,-1]
])

# ==========================================================
# FONCTION DE CONVOLUTION
# ==========================================================

# Cette fonction applique le filtre sur l'image.
def convolution(image, kernel):

    # Récupérer la hauteur et la largeur de l'image.
    h, w = image.shape

    # Récupérer la hauteur et la largeur du filtre.
    kh, kw = kernel.shape

    # Calculer la hauteur de la sortie.
    output_h = h - kh + 1

    # Calculer la largeur de la sortie.
    output_w = w - kw + 1

    # Créer une matrice remplie de zéros.
    # Elle contiendra le résultat de la convolution.
    output = np.zeros((output_h, output_w))

    # Parcourir toutes les lignes possibles.
    for i in range(output_h):

        # Parcourir toutes les colonnes possibles.
        for j in range(output_w):

            # Extraire une région de l'image
            # ayant la même taille que le filtre.
            region = image[i:i+kh, j:j+kw]

            # Multiplier chaque pixel par le filtre
            # puis additionner les résultats.
            output[i, j] = np.sum(region * kernel)

    # Retourner la carte de caractéristiques.
    return output

# ==========================================================
# FONCTION ReLU
# ==========================================================

# Fonction d'activation.
# Les valeurs négatives deviennent 0.
def relu(x):

    return np.maximum(0, x)

# ==========================================================
# MAX POOLING
# ==========================================================

# Réduction de la taille de la carte de caractéristiques.
def max_pooling(feature_map, size=2):

    # Dimensions de la carte.
    h, w = feature_map.shape

    # Dimensions de la sortie.
    output_h = h // size
    output_w = w // size

    # Création de la matrice résultat.
    output = np.zeros((output_h, output_w))

    # Parcours des blocs.
    for i in range(output_h):

        for j in range(output_w):

            # Extraire un bloc 2×2.
            region = feature_map[
                i*size:(i+1)*size,
                j*size:(j+1)*size
            ]

            # Conserver la valeur maximale.
            output[i, j] = np.max(region)

    # Retourner la matrice réduite.
    return output

# ==========================================================
# FLATTEN
# ==========================================================

# Transformer la matrice en un seul vecteur.
def flatten(x):

    return x.flatten()

# ==========================================================
# COUCHE DENSE
# ==========================================================

# Couche entièrement connectée.
def dense(x, weights, bias):

    # Calcul :
    # y = xW + b
    return np.dot(x, weights) + bias

# ==========================================================
# PROGRAMME PRINCIPAL
# ==========================================================

# Appliquer la convolution.
feature_map = convolution(image, kernel)

# Afficher la carte de caractéristiques.
print("Après convolution :")
print(feature_map)

# Appliquer la fonction ReLU.
feature_map = relu(feature_map)

# Afficher le résultat.
print("\nAprès ReLU :")
print(feature_map)

# Réduire la taille avec MaxPooling.
feature_map = max_pooling(feature_map)

# Afficher le résultat.
print("\nAprès MaxPooling :")
print(feature_map)

# Transformer la matrice en vecteur.
x = flatten(feature_map)

# Afficher le vecteur.
print("\nVecteur :")
print(x)

# Générer aléatoirement les poids.
weights = np.random.rand(len(x), 2)

# Générer les biais.
bias = np.random.rand(2)

# Calcul de la sortie finale.
output = dense(x, weights, bias)

# Afficher les scores obtenus.
print("\nSortie du réseau :")
print(output)

# Chercher la classe ayant le score maximal.
prediction = np.argmax(output)

# Afficher la classe prédite.
print("\nClasse prédite :", prediction)
```
