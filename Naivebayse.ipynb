```python
import numpy as np  # Importation de NumPy pour les calculs numériques

class NaiveBayesScratch:
    def fit(self, X, y):
        # Récupérer toutes les classes uniques
        self.classes = np.unique(y)

        # Dictionnaires pour stocker les paramètres
        self.mean = {}   # moyenne par classe
        self.var = {}    # variance par classe
        self.priors = {} # probabilité a priori de chaque classe

        # Parcourir chaque classe
        for c in self.classes:
            # Extraire les données appartenant à la classe c
            X_c = X[y == c]

            # Calcul de la moyenne de chaque feature pour la classe c
            self.mean[c] = np.mean(X_c, axis=0)

            # Calcul de la variance de chaque feature pour la classe c
            self.var[c] = np.var(X_c, axis=0)

            # Probabilité a priori P(C)
            self.priors[c] = len(X_c) / len(X)

    def _gaussian(self, x, mean, var):
        # Petite valeur pour éviter division par zéro
        eps = 1e-6

        # Formule de la distribution gaussienne
        coeff = 1 / np.sqrt(2 * np.pi * var + eps)

        # Partie exponentielle de la formule
        exponent = np.exp(-((x - mean) ** 2) / (2 * var + eps))

        # Probabilité finale
        return coeff * exponent

    def predict(self, X):
        results = []  # liste des prédictions

        # Pour chaque échantillon à prédire
        for x in X:
            posteriors = []  # stocker les probabilités par classe

            # Tester chaque classe possible
            for c in self.classes:

                # Log de la probabilité a priori P(C)
                prior = np.log(self.priors[c])

                # Calcul de la vraisemblance P(X|C)
                likelihood = np.sum(
                    np.log(self._gaussian(x, self.mean[c], self.var[c]))
                )

                # Probabilité finale (Bayes)
                posterior = prior + likelihood

                posteriors.append(posterior)

            # Choisir la classe avec la plus grande probabilité
            results.append(self.classes[np.argmax(posteriors)])

        return np.array(results)  # retourner les prédictions
```

```python


# =========================
#  Données d'entraînement
# =========================
X = np.array([
    [1, 2], [2, 1], [2, 3],  # classe 0
    [6, 7], [7, 8], [8, 6]   # classe 1
])

y = np.array([0, 0, 0, 1, 1, 1])

# =========================
#  Création du modèle
# =========================
model = NaiveBayesScratch()

# =========================
#  Entraînement
# =========================
model.fit(X, y)

# =========================
#  Données test
# =========================
X_test = np.array([[2, 2], [7, 7]])

# =========================
#  Prédiction
# =========================
print("Prédictions :", model.predict(X_test))
```
